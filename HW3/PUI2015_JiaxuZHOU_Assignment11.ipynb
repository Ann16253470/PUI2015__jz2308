```python
from __future__ import print_function
__author__= 'jzhou'
import os

from scipy.optimize import curve_fit, minimize
import numpy as np
import pylab as pl
import json
%pylab inline
s = json.load( open('fbb_matplotlibrc.json') )
pl.rcParams.update(s)


from scipy.stats import norm
import matplotlib.mlab as mlab
```

```python
mymean=100
df=100
mysize=2000/(np.array(range(1,100)+[10]))
```

```python
md={}
means = {}
```

###$\chi^2$Distribution

```python
np.random.seed(10)
md['chisq'] = np.random.chisquare(df, size=100)
md['chisq']={}
df=100

for n in mysize:
    md['chisq'][n] = np.random.chisquare(df, size=n)
    

means['chisq'] = {}
fig_mu_N = pl.figure(figsize=(10,10))
axchisq_mu_n = fig_mu_N.add_subplot(111)



for nn in md['chisq'].iterkeys():
    means['chisq'][nn] = md['chisq'][nn].mean()
    #plot it
    axchisq_mu_n.plot(nn, means['chisq'][nn], 'o')
    axchisq_mu_n.set_xlabel('sample size')
    axchisq_mu_n.set_ylabel('sample mean')
    axchisq_mu_n.plot([min(mysize), max(mysize)],[df,df], 'k')
```

```python
allmeans=[]
for n in means['chisq']:
    allmeans.append(means['chisq'][n])
pl.figure(figsize=(10,10))

(mu, sigma) = norm.fit(allmeans)
n, bins, patches = plt.hist(allmeans,bins=30, normed=1, facecolor='green', alpha=0.75)
 
y = mlab.normpdf(bins, mu, sigma)
l = pl.plot(bins, y, 'r--', linewidth=2)

pl.xlabel('sample mean')
pl.ylabel('N')
```

###Normal Distribution

```python
np.random.seed(10)
md['normal'] = np.random.normal(df, size=100)
md['normal']={} 

for n in mysize:
    md['normal'][n] = np.random.normal(mymean,df,size=n)
    
means['normal'] = {}
fig_mu_N = pl.figure(figsize=(10,10))
axchisq_mu_n = fig_mu_N.add_subplot(111)
for nn in md['normal'].iterkeys():
    means['normal'][nn] = md['normal'][nn].mean()
    #plot it
    axchisq_mu_n.plot(nn, means['normal'][nn], 'o')
    axchisq_mu_n.set_xlabel('sample size')
    axchisq_mu_n.set_ylabel('sample mean')
    axchisq_mu_n.plot([min(mysize), max(mysize)],[df,df], 'k')
```

```python
allmeans=[]
for n in means['normal']:
    allmeans.append(means['normal'][n])
pl.figure(figsize=(10,10))

pl.figure(figsize=(10,10))

(mu, sigma) = norm.fit(allmeans)
n, bins, patches = plt.hist(allmeans,bins=30, normed=1, facecolor='green', alpha=0.75)
 
y = mlab.normpdf(bins, mu, sigma)
l = pl.plot(bins, y, 'r--', linewidth=2)

pl.xlabel('sample mean')
pl.ylabel('N')
```

###Poisson Distribution

```python
np.random.seed(10)
md['poisson']={}
for n in mysize:
    md['poisson'][n] = np.random.poisson(mymean,size=n)
    
means['poisson'] = {}
fig_mu_N = pl.figure(figsize=(10,10))
axchisq_mu_n = fig_mu_N.add_subplot(111)
for nn in md['poisson'].iterkeys():
    means['poisson'][nn] = md['poisson'][nn].mean()
    axchisq_mu_n.plot(nn, means['poisson'][nn], 'o')
    axchisq_mu_n.set_xlabel('sample size')
    axchisq_mu_n.set_ylabel('sample mean')
axchisq_mu_n.plot([min(mysize), max(mysize)],[mymean,mymean], 'k')
```

```python
allmeans=[]
for n in means['poisson']:
    allmeans.append(means['poisson'][n])
pl.figure(figsize=(10,10))
pl.figure(figsize=(10,10))

(mu, sigma) = norm.fit(allmeans)
n, bins, patches = plt.hist(allmeans,bins=30, normed=1, facecolor='green', alpha=0.75)
 
y = mlab.normpdf(bins, mu, sigma)
l = pl.plot(bins, y, 'r--', linewidth=2)

pl.xlabel('sample mean')
pl.ylabel('N')
```
